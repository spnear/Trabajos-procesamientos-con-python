# Total de homicidios desde 1995 hasta el año 2000

# homicidios desde el año 1995 hasta el año 2000 perpetrado por hombres de raza negra por sofocacion=Suffocation

```python
dataset = homicidios[(homicidios["Year"] >= 1995) & (homicidios["Year"] <= 2000)]
# (dataset["Perpetrator Race"] == "Black").value_counts() # Cantidad por raza negra
# dataset[(dataset["Perpetrator Race"] == "Black") & (dataset["Weapon"] == "Suffocation")] # Casos por sofocación
dataset[(dataset["Perpetrator Race"] == "Black") & (dataset["Weapon"] == "Suffocation")]["Agency Name"].value_counts().sum()
```
