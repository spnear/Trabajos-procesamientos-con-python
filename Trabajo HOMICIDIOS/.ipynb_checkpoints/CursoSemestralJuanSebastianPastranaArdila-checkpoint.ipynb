```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# Consumir base de datos.
db = pd.read_csv("database.csv")
```

```python
db.info()
```

# 1° 5 Primeras ciudades con el mayor número de agencias

```python
# 5 Primeras ciudades con el mayor número de agencias
db['City'].value_counts().head(5)
```

```python
db['City'].value_counts().head(5).plot(kind = 'bar')
```

# 2° De los tipos de agencia, determinar las de tipo Sheriff, y mostrar el nombre 

```python
tipoAgencia = "Sheriff"
db[db["Agency Type"]==tipoAgencia]["Agency Name"].value_counts().sort_values(ascending = False)
```

 # 3° Listar los estados mas afectados por crimenes perpetrados por mujeres 

```python
perpetradorSex = "Female"
db[db["Perpetrator Sex"]==perpetradorSex]["State"].value_counts().sort_values(ascending = False)
```

```python
db[db["Perpetrator Sex"]==perpetradorSex]["State"].value_counts().plot(kind='bar',figsize=(100,50),legend = True,fontsize=100)
```

 # 4° Listar los estados mas afectados por crimenes perpetrados por hombres 

```python
perpetradorSex = "Male"
db[db["Perpetrator Sex"]==perpetradorSex]["State"].value_counts().sort_values(ascending = False)
```

```python
db[db["Perpetrator Sex"]==perpetradorSex]["State"].value_counts().plot(kind='bar',figsize=(100,50),legend = True,fontsize=60)
```

 # 5° Determinar el numero exacto del numero de crimenes hechos por mujeres de raza Asian/Pacific Islander 

```python
db[(db["Perpetrator Race"]=="Asian/Pacific Islander") & (db["Perpetrator Sex"]=="Female")].count().max()
```

```python
db["Perpetrator Race"]
#db[(db["Perpetrator Race"]=="Asian/Pacific Islander") & (db["Perpetrator Sex"]=="Female")].count().plot(figsize=(10,5),fontsize=10,kind="bar")
```

# 6° Determinar la raza más criminal

```python
db["Perpetrator Race"].max()
```

# 7° Determinar el numero exacto de hispanos que han asesinado mediante la estrangulacion 

```python
db[(db["Perpetrator Ethnicity"]=="Hispanic") & (db["Weapon"]=="Strangulation")].count().max()
```

# 8° Determinar el tipo de relación más peligrosa, el cual comete más homicidios con armas de tipo Shotgun=escopeta 

```python
db[db["Weapon"]=="Shotgun"]["Relationship"].sort_values().value_counts().head(1)
```

```python
db[db["Weapon"]=="Shotgun"]["Relationship"].sort_values().value_counts().plot(kind = 'bar',figsize=(50,20),fontsize=50)
```

# 9°Cual es el sexo que mas homicidios han cometido con Veneno= Poison 

```python
db[db["Weapon"]=="Poison"]["Perpetrator Sex"].sort_values().value_counts().head(1)
```

```python
db[db["Weapon"]=="Poison"]["Perpetrator Sex"].sort_values().value_counts().plot(kind = 'barh',figsize=(5,5),fontsize=10)
```

# 10° Cuantos asesinos de raza negra atrapo el FBI 

```python
db[(db["Record Source"]=="FBI") & (db["Perpetrator Race"]=="Black")].count().max()
```

```python
db[(db["Record Source"]=="FBI") & (db["Perpetrator Race"])].plot(kind = 'bar',figsize=(5,5),fontsize=10)
```

# 11° Cuantas victimas de tipo hispanas han muerto por armas de tipo explosivo=Explosives 

```python
db[(db["Perpetrator Ethnicity"]=="Hispanic") & (db["Weapon"]=="Explosives")].count().max()
```

# 12° Cual ha sido el asesino más viejo

# 14° Cual es el total de homicidios desde el año 1995 hasta el año 2000 

```python
db[(db["Year"]>=1995) & (db["Year"]<=2000)].count().max()
```
