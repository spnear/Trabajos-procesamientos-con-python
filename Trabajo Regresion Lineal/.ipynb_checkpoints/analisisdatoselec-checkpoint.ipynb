# MATRICES DE ENTRANIMIENTO Y PRUEBA


```python
x = datosplanta[['Temperature', 'Exhaust Vacuum', 'Ambient Presure','Relative Humidity']]

y = datosplanta["PE"]  # Variable a predecir 
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, Y_train)
```

Evaluación del modelo

```python
print(lm.intercept_)
```

```python
print (lm.coef_)
```

```python
coeff = pd.DataFrame(lm.coef_,x.columns,columns=['Coeficientes'])
coeff
```

```python
predictic = lm.predict(X_test)
```

```python
predictic
```

```python
Y_test.head()
```

```python
plt.scatter(Y_test,predictic)
```

```python
sns.distplot(Y_test-predictic)
```

```python
from sklearn import metrics
```

```python
print('MAE', metrics.mean_absolute_error(Y_test, predictic))
print('MSE', metrics.mean_squared_error(Y_test, predictic))
print('RMSE', np.sqrt(metrics.mean_squared_error(Y_test, predictic)))
```

```python
predictionOut = widgets.Output()
```

```python
def obtener():
    dato=(a.value,b.value,c.value,d.value)
    return dato
```

```python
def predecir(b):
    dato = obtener()
    prediction = lm.predict([dato])
    
    #limpiamos el area del aoutput antes de imprimir un resultado
    predictionOut.clear_output(wait=False)
    with predictionOut:
        print("Resultado ")
        print("------------------------------------------------------------------------------")
        print("Datos: [Temperature, Exhaust Vacuum, Ambient Presure, 'Relative Humidity]")
        print("Prediccion = ", prediction[0], "MW consumidos de energía")
        print("------------------------------------------------------------------------------")
```
